### Limpieza de Datos

Tenemos dos alternativas para la limpieza de datos, una destructiva y otra generativa.
Cuando hablamos de la opción destructiva, nos referimos al eliminado de datos

```python
rows_without_nan =  df.dropna(axis='index')
columns_without_nan = df.dropna(axis=1)
```

```python
columns_without_nan
```

La alternativa generativa consiste en poblar de datos los valores nulos sim importar la fila o la columna. Cómo recomendación se debe trabajar por columnas individuales.

```python
# converted = df.fillna(0) NO HACER, NO COLOCAR 0 AUNQUE SEA UNA PRÁCTICA COMÚN
df.seasons = df.seasons.fillna(df.seasons.mean())
df.imdb_score = df.imdb_score.fillna(df.imdb_score.mean()) #5.970558
```

```python
df.tmdb_score = df.tmdb_score.fillna(df.tmdb_score.mean())
df.describe()
```
